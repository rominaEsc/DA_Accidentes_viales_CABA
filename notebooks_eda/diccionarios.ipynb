# Carga de datos

```python
df_diccionario_homicidios = pd.read_excel(io='../data/homicidios.xlsx',sheet_name='DICCIONARIO_HECHOS')
df_diccionario_victimas = pd.read_excel(io='../data/homicidios.xlsx',sheet_name='DICCIONARIO_VICTIMAS')
```

----

# Archivo 'homicidios.xlsx', sheet_name='DICCIONARIO_HECHO

Extracción de un diccionario de variables del df homicidios 


```python
df_diccionario_homicidios
```

```python
diccionario_variables_homicidios = df_diccionario_homicidios[['variables y definiciones','descripción']].head(21)
```

```python
diccionario_variables_homicidios.columns=['variables_y_definiciones', 'descripcion']
```

```python
diccionario_variables_homicidios
```

```python
diccionario_variables_homicidios.variables_y_definiciones = diccionario_variables_homicidios.variables_y_definiciones.replace({r'XY (CABA)': 'XY'})
diccionario_variables_homicidios.variables_y_definiciones = diccionario_variables_homicidios.variables_y_definiciones.replace({r'ó':'o',r' ':'_'},regex=True)
diccionario_variables_homicidios.variables_y_definiciones = diccionario_variables_homicidios.variables_y_definiciones.str.lower()

diccionario_variables_homicidios.descripcion = diccionario_variables_homicidios.descripcion.str.capitalize()
```

```python
diccionario_variables_homicidios
```

```python
diccionario_variables_homicidios.to_csv('../data/diccionarios/diccionario_variables_del_df_homicidios.csv',index=False)
```

----

Extracción de los valores que pueden tomar las variables Tipo de calle, victima, acusado y participantes.

```python
df_variables_y_definiciones = df_diccionario_homicidios.tail(23)
df_variables_y_definiciones.columns = ['variables','valores','descripcion']
df_variables_y_definiciones = df_variables_y_definiciones.reset_index(drop=True)
df_variables_y_definiciones['variables'].fillna(method='ffill', inplace=True)
df_variables_y_definiciones.descripcion = df_variables_y_definiciones.descripcion.str.capitalize()
```

```python
df_variables_y_definiciones
```

```python
diccionario_variables_homicidios.to_csv('../data/diccionarios/diccionario_valores_de_variables_del_df_homicidios.csv',index=False)
```

# Archivo 'data/homicidios.xlsx',sheet_name='DICCIONARIO_VICTIMAS'

* Extracción de los valores que pueden tomar las variables Tipo de calle, victima, acusado y participantes.

```python
df_diccionario_victimas
```

```python
diccionario_victimas_variables_y_definiciones = df_diccionario_victimas.head(9)
diccionario_victimas_variables_y_definiciones = diccionario_victimas_variables_y_definiciones.drop(columns=['Unnamed: 2'])
diccionario_victimas_variables_y_definiciones.columns = ['variables y definiciones','descripcion']
diccionario_victimas_variables_y_definiciones.descripcion = diccionario_victimas_variables_y_definiciones.descripcion.str.capitalize()
diccionario_victimas_variables_y_definiciones
```

```python
diccionario_victimas_variables_y_definiciones.to_csv('../data/diccionarios/diccionario_de_variables_y_definiciones_del_df_victimas.csv',index=False)
```

---

* Extracción de la clasificación de las víctimas

```python
df_diccionario_victimas.tail(9)
victima_clasificacion = df_diccionario_victimas.tail(9)
victima_clasificacion = victima_clasificacion.drop(columns='variables y definiciones')
victima_clasificacion.columns = ['valores','descripcion']
victima_clasificacion.descripcion = victima_clasificacion.descripcion.str.capitalize()
victima_clasificacion = victima_clasificacion.reset_index(drop=True)
victima_clasificacion
```

```python
victima_clasificacion.to_csv('../data/diccionarios/diccionario_clasificacion_victimas_del_df_victimas.csv',index=False)
```
